```python
from geopy.geocoders import Nominatim
import time
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
tqdm.pandas()
```

```python
# instantiate a new Nominatim client
app = Nominatim(user_agent="tutorial")
```

```python
def get_location_by_address(address):
    """This function returns a location as raw from an address
    will repeat until success"""
    time.sleep(1)
    try:
        return app.geocode(address).raw
    except:
        return get_location_by_address(address)
```

```python
def get_location_by_address(address):
    try:
        return app.geocode(address).raw
    except:
        return get_location_by_address(address)
```

```python
df = pd.read_csv ('clean_address_13.csv')
```

```python
df['full_address'] = df['clean_address2'] + ', ' + df['Suburb'] + ' ' + df['PostCode'].astype(str)
df.head()
```

```python
latitude_list = []
longitutde_list = []
for location_name in tqdm(df.full_address):
    print(location_name)
    if "/" in str(location_name):
        latitude_list.append("check")
        longitutde_list.append("check")
    else:
        location = get_location_by_address(location_name) 
        latitude_list.append(location["lat"])
        longitutde_list.append(location["lon"])
df['Latitude'] = latitude_list
df['Longitude'] = longitutde_list
```

```python
df.to_csv('clean_address_13&.csv')
```
