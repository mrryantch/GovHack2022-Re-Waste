```python
import pandas as pd
import numpy as np
```

```python
def squeeze_nan(x):
        original_columns = x.index.tolist()
        squeezed = x.dropna()
        squeezed.index = [original_columns[n] for n in range(squeezed.count())]
        return squeezed.reindex(original_columns, fill_value=np.nan)
```

```python
df = pd.read_csv ('business_location.csv')
temp_df = df.StreetAddress.str.split(" ",expand=True)
temp_df= temp_df[temp_df.columns[::-1]]
temp_df = temp_df.apply(squeeze_nan, axis=1)
temp_df
```

```python
temp_df[8]= temp_df[8].astype(str).replace(['Ground','nan', '/'],'')
df['clean_address'] = temp_df[8]+ ' ' + temp_df[9] + ' ' + temp_df[10] + ', ' + df['Suburb'] + ' ' + df['PostCode'].astype(str)
df.head()
```

```python
df.to_csv('clean_address.csv')
```
